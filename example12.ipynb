```python
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",hue="sex", size="size", sizes=(50,200), data=tips)
plt.xlabel("Total bill")
plt.ylabel("tips")
plt.title("Relationship between tips and total bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri=sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",hue="event",style="region",markers=True,dashes=False,data=fmri)
plt.xlabel("Timepoint")
plt.ylabel("signal Intensity")
plt.title("changes of signal Intensity over time")
plt.show()
```

```python
titanic=sns.load_dataset("titanic")
sns.barplot(x="class",y="fare",data=titanic)
```

```python
iris=sns.load_dataset("iris")
sns.histplot(data=iris, x="petal_length",bins=20,kde=True,color="green")
plt.xlabel("petal length")
plt.ylabel("Frequency")
plt.title("Distribution of petal lengths in Iris Flowers")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
diamonds=sns.load_dataset("diamonds")
sns.scatterplot(x="carat",y="price",data=diamonds)
plt.text(1,1800,"Large,Expensive diamonds",fontsize=12,color="red")
plt.text(2.5,5000,"Small,Expensive diamonds",fontsize=12,color="blue")
plt.title("diamonds cost by carat")
plt.xlabel("Carat (ct)")
plt.ylabel("Price ($)")
plt.show()
```
